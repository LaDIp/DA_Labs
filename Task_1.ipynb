#Задача №2

```
data = pd.read_csv('weather.csv', sep=';', comment='#')
data.head()
```

```
fig, axes = plt.subplots(figsize=(15,5))
axes.plot(data['time'], data['T'])
axes.set_title("Температура воздуха от времени")
plt.setp(plt.xticks()[1], rotation=45)
fig.show()
```

```
fig, axes = plt.subplots(figsize=(15,5))
axes.plot(data['time'], data['P0'])
axes.set_title("Атмосферное давление от времени")
plt.setp(plt.xticks()[1], rotation=45)
fig.show()
```

```
fig, axes = plt.subplots(figsize=(15,5))
axes.plot(data['time'], data['U'])
axes.set_title("Относительная влажность от времени")
plt.setp(plt.xticks()[1], rotation=45)
fig.show()
```

#Задание №3

```
wind = data['DD']
wind.value_counts()
```

```
ne = wind[wind == 'СЗ'].count() + wind[wind == 'П'].count()
nw = wind[wind == 'СB'].count() + wind[wind == 'П'].count()
sw = wind[wind == 'ЮВ'].count() + wind[wind == 'П'].count()
w = wind[wind == 'В'].count() + wind[wind == 'П'].count()
n = wind[wind == 'С'].count() + wind[wind == 'П'].count()

r = np.array([w, nw, n, ne, sw, w])

theta = [0, np.pi*0.25, np.pi*0.5 , np.pi*0.75, np.pi*1.75, 0]

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)
ax.set_rticks(range(0, 40, 10))

plt.show()
```
